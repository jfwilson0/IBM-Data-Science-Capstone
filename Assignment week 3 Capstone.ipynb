```python
import urllib.request
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
page = urllib.request.urlopen(url)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page, "lxml")
```

```python
all_tables=soup.find_all("table")
```

```python
right_table=soup.find('table', class_='wikitable sortable')
```

```python
A=[]
B=[]
C=[]

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

```python
import pandas as pd
df=pd.DataFrame(A,columns=['Postal code'])
df['Borough']=B
df['Neighborhood']=C
df
```

```python
df = df[(df.Borough!="Not assigned\n")]
df
```

```python
df['Postal code'] = df['Postal code'].astype(str).str[:-1]
df['Borough'] = df['Borough'].astype(str).str[:-1]
df['Neighborhood'] = df['Neighborhood'].astype(str).str[:-1]
df
```

```python
df
```

```python
strJoin = lambda x:",".join(x.astype(str))     
newdf = df.groupby('Postal code').agg({"Borough":strJoin,"Neighborhood":strJoin})
```
